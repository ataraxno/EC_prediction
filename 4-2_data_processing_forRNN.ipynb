```python
columns = ["ECD(1)", "ECD(2)", "ECD(3)", "ECD(4)", 'ECS(1)', 'ECS(2)', 'ECS(3)', 'ECS(4)', \
           'DrgV(1)', 'DrgV(2)', 'DrgV(3)', 'DrgV(4)', \
           'Mixlevel(1)', 'Mixlevel(2)', 'Mixlevel(3)', 'Mixlevel(4)', 'StklevelA(1)', 'StklevelA(2)', 'StklevelA(3)', 'StklevelA(4)', \
           'StklevelB(1)', 'StklevelB(2)', 'StklevelB(3)', 'StklevelB(4)',\
           'RWMC(1)', 'RWMC(2)', 'RWMC(3)', 'RWMC(4)', "RWEC(1)", "RWEC(2)", "RWEC(3)", "RWEC(4)",\
           'InjDrgVR(1)', 'InjDrgVR(2)', 'InjDrgVR(3)', 'InjDrgVR(4)', \
           'InjWtrVR(1)', 'InjWtrVR(2)', 'InjWtrVR(3)', 'InjWtrVR(4)', 'InjStkVR(1)', 'InjStkVR(2)', 'InjStkVR(3)', 'InjStkVR(4)', \
           'StkVA(1)', 'StkVA(2)', 'StkVA(3)', 'StkVA(4)', 'StkVB(1)', 'StkVB(2)', 'StkVB(3)', 'StkVB(4)', \
           'DrgDaily(1)', 'DrgDaily(2)', 'DrgDaily(3)', 'DrgDaily(4)', 'MixV(1)', 'MixV(2)', 'MixV(3)', 'MixV(4)', \
           'mLperDrip', 'CO2', 'Pyrano_W', 'AccdRadSet', 'IrrDaily', 'PlantH', 'NodeNum','temp','hum']
```

```python
raw_df = raw_df[columns]
raw_df
```

```python
index = np.arange(time_step, raw_df.shape[0]-output_size)
np.random.shuffle(index)
```

```python
temp_label = raw_label.iloc[index[0]:index[0]+output_size,:]
temp_label = temp_label.as_matrix().T
temp_label = temp_label.reshape(-1, 4, output_size)

temp_data = raw_df.iloc[index[0]-time_step:index[0],:]

_ = 1

for i in index:
    if i == index[0]:
        continue

    temp_label = np.append(temp_label, raw_label.iloc[i:i+output_size, :].as_matrix().T.reshape(-1,4,output_size), axis=0)
    temp_data = temp_data.append(raw_df.iloc[i-time_step:i,:])
    
    _ = _ + 1
    
    if _ % 1000 == 0:
        print("%5d/%5d processing..." % (_, raw_df.shape[0]-output_size-time_step))
    
print("%5d/%5d process ended." % (_, raw_df.shape[0]-output_size-time_step))

raw_data = temp_data
raw_label = temp_label
```

```python
#Processing
temp = []
mins = []
maxs = []
labelmins = []
labelmaxs = []

for i in range(raw_data.shape[1]):
    temp = raw_data[:,i]
    min_ = min(temp)
    max_ = max(temp)
    temp = (temp - min_) / (max_ - min_)
    raw_data[:,i] = temp
    mins.append(min_)
    maxs.append(max_)

    
for i in range(raw_label.shape[1]):
    temp = raw_label[:,i]
    labelmin_ = min(temp)
    labelmax_ = max(temp)
    temp = (temp - labelmin_) / (labelmax_ - labelmin_)
    raw_label[:,i] = temp
    labelmins.append(labelmin_)
    labelmaxs.append(labelmax_)
```

```python
train_label = raw_label[:TRAIN,:]
train_data = raw_data[:TRAIN*TIME_STEP,:]

test_label = raw_label[TRAIN:,:]
test_data = raw_data[TRAIN*TIME_STEP:,:]
```

```python
print(train_data.shape)
print(train_label.shape)

print(test_data.shape)
print(test_label.shape)
```

```python
f = open("./Data/ECtrainset_24.npz", 'wb')
np.savez(f, traindataset = train_data, trainlabel = train_label, testdataset = test_data, testlabel = test_label, \
                 mins = mins, maxs = maxs, labelmins = labelmins, labelmaxs = labelmaxs)
f.close()
```

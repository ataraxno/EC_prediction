```python
import pandas as pd
import numpy as np

print("Package is ready.")
```

```python
raw_df = pd.read_csv("./Data/EC_1hour_zipped.dat", index_col=["Unnamed: 0"])
```

```python
raw_df = pd.DataFrame(raw_df, dtype="float")
raw_df = raw_df.iloc[360:,:]
raw_df
```

```python
raw_df.columns.tolist()
```

```python
raw_label = raw_df[["RWEC(1)","RWEC(2)","RWEC(3)","RWEC(4)"]]
raw_label
```

```python
columns = ["ECD(1)", "ECD(2)", "ECD(3)", "ECD(4)", 'ECS(1)', 'ECS(2)', 'ECS(3)', 'ECS(4)', \
           'DrgV(1)', 'DrgV(2)', 'DrgV(3)', 'DrgV(4)', \
           'Mixlevel(1)', 'Mixlevel(2)', 'Mixlevel(3)', 'Mixlevel(4)', 'StklevelA(1)', 'StklevelA(2)', 'StklevelA(3)', 'StklevelA(4)', \
           'StklevelB(1)', 'StklevelB(2)', 'StklevelB(3)', 'StklevelB(4)',\
           'RWMC(1)', 'RWMC(2)', 'RWMC(3)', 'RWMC(4)', "RWEC(1)", "RWEC(2)", "RWEC(3)", "RWEC(4)",\
           'InjDrgVR(1)', 'InjDrgVR(2)', 'InjDrgVR(3)', 'InjDrgVR(4)', \
           'InjWtrVR(1)', 'InjWtrVR(2)', 'InjWtrVR(3)', 'InjWtrVR(4)', 'InjStkVR(1)', 'InjStkVR(2)', 'InjStkVR(3)', 'InjStkVR(4)', \
           'StkVA(1)', 'StkVA(2)', 'StkVA(3)', 'StkVA(4)', 'StkVB(1)', 'StkVB(2)', 'StkVB(3)', 'StkVB(4)', \
           'DrgDaily(1)', 'DrgDaily(2)', 'DrgDaily(3)', 'DrgDaily(4)', 'MixV(1)', 'MixV(2)', 'MixV(3)', 'MixV(4)', \
           'mLperDrip', 'CO2', 'Pyrano_W', 'AccdRadSet', 'IrrDaily']
```

```python
raw_df = raw_df[columns]
raw_df
```

```python
time_step = 30 # A day + A quater
output_size = 18 # 3/4 day

index = np.arange(time_step, raw_df.shape[0]-output_size)
np.random.shuffle(index)
```

```python
temp_label = raw_label.iloc[index[0]:index[0]+output_size,:]
temp_label = temp_label.as_matrix().T
temp_label = temp_label.reshape(-1, 4, output_size)

temp_data = raw_df.iloc[index[0]-time_step:index[0],:]

_ = 1

for i in index:
    if i == index[0]:
        continue

    temp_label = np.append(temp_label, raw_label.iloc[i:i+output_size, :].as_matrix().T.reshape(-1,4,output_size), axis=0)
    temp_data = temp_data.append(raw_df.iloc[i-time_step:i,:])
    
    _ = _ + 1
    
    if _ % 1000 == 0:
        print("%5d/%5d processing..." % (_, raw_df.shape[0]-output_size-time_step))
    
print("%5d/%5d process ended." % (_, raw_df.shape[0]-output_size-time_step))

raw_data = temp_data
raw_label = temp_label
```

```python
choice = np.array([x for x in range(0, 60, 4)])
plus = np.array([x for x in range(60,65)])


choice_1 = np.append(choice, plus)
train_df_1 = raw_data.iloc[:,choice_1]

choice_2 = np.append(choice + 1, plus)
train_df_2 = raw_data.iloc[:,choice_2]

choice_3 = np.append(choice + 2, plus)
train_df_3 = raw_data.iloc[:,choice_3]

choice_4 = np.append(choice + 3, plus)
train_df_4 = raw_data.iloc[:,choice_4]
```

```python
train_df_4
```

```python
train_label_1 = train_df_1["ECD(1)"]
del train_df_1["ECD(1)"]

train_label_2 = train_df_2["ECD(2)"]
del train_df_2["ECD(2)"]

train_label_3 = train_df_3["ECD(3)"]
del train_df_3["ECD(3)"]

train_label_4 = train_df_4["ECD(4)"]
del train_df_4["ECD(4)"]
```

```python
train_df_1 = train_df_1.as_matrix()
train_label_1 = train_label_1.as_matrix()
train_label_1 = train_label_1.reshape(-1, 1)

train_df_2 = train_df_2.as_matrix()
train_label_2 = train_label_2.as_matrix()
train_label_2 = train_label_2.reshape(-1, 1)

train_df_3 = train_df_3.as_matrix()
train_label_3 = train_label_3.as_matrix()
train_label_3 = train_label_3.reshape(-1, 1)

train_df_4 = train_df_4.as_matrix()
train_label_4 = train_label_4.as_matrix()
train_label_4 = train_label_4.reshape(-1, 1)
```

```python
train_data = np.concatenate((train_df_1, train_df_2, train_df_3, train_df_4), axis=0)
train_label = np.concatenate((train_label_1, train_label_2, train_label_3, train_label_4), axis=0)
```

```python
print(train_data.shape)
print(train_label.shape)
```

```python
np.savetxt("./Data/EC_train_fMLP.csv", train_data, delimiter=",")
np.savetxt("./Data/EC_train_label_fMLP.csv", train_label, delimiter=",")
```
